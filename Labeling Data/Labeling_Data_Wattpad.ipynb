```python
import pandas as pd

# Load data 1000 baris
df = pd.read_csv('/content/drive/MyDrive/labeled_data.csv')

# Pisahkan data
labeled = df[df['label'].notna()]
unlabeled = df[df['label'].isna()]

# Cek jumlah
print("Jumlah data labeled :", len(labeled))
print("Jumlah data unlabeled :", len(unlabeled))

# Simpan ke file
labeled.to_csv('data_labeled_250.csv', index=False)
unlabeled.to_csv('data_unlabeled_750.csv', index=False)

print("Selesai: 250 labeled & 750 unlabeled sudah dipisah!")
```

```python
import pandas as pd
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Load 250 data
df = pd.read_csv('data_labeled_250.csv')

X_train = df['content']
y_train = df['label']

# TF-IDF Vectorizer
vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)

# Training Logistic Regression
model = LogisticRegression(max_iter=300)
model.fit(X_train_tfidf, y_train)

# Simpan model + vectorizer
pickle.dump(model, open('sentiment_model.pkl', 'wb'))
pickle.dump(vectorizer, open('tfidf.pkl', 'wb'))

print("Training selesai! Model = sentiment_model.pkl, Vectorizer = tfidf.pkl")
```

```python
import pandas as pd
import pickle

# Data
df = pd.read_csv('data_unlabeled_750.csv')

# Model
model = pickle.load(open('sentiment_model.pkl', 'rb'))
vectorizer = pickle.load(open('tfidf.pkl', 'rb'))

# TFIDF
X_tfidf = vectorizer.transform(df['content'])

# Prediksi
prediksi = model.predict(X_tfidf)

df['label'] = prediksi

# Simpan
df.to_csv('data_hasil_labeling.csv', index=False)

print("Label otomatis selesai! Hasil = data_hasil_labeling.csv")
```
